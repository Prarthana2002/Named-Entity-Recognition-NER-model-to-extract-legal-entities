```python
import os
from pdf2image import convert_from_path
from PIL import Image
import pytesseract
import time

# Install Poppler and Tesseract if not done already
!apt-get install -y poppler-utils tesseract-ocr
!pip install pytesseract pdf2image Pillow

# Folders
input_folder = '/content/PDF'          # Your uploaded PDFs
output_folder = '/content/text output' # Output text
os.makedirs(output_folder, exist_ok=True)

# Start processing
start_time = time.time()

for filename in os.listdir(input_folder):
    if filename.lower().endswith('.pdf'):
        pdf_path = os.path.join(input_folder, filename)
        base_name = os.path.splitext(filename)[0]
        print(f"\nüîç Processing: {filename}")

        try:
            images = convert_from_path(pdf_path, dpi=300)
        except Exception as e:
            print(f"‚ùå Error converting {filename}: {e}")
            continue

        full_text = ""

        for i, image in enumerate(images):
            # Preprocess for better OCR
            gray = image.convert('L')
            threshold = gray.point(lambda x: 0 if x < 150 else 255, '1')

            # OCR configuration
            config = '--oem 3 --psm 1'
            text = pytesseract.image_to_string(threshold, lang='eng', config=config)

            full_text += f"\n\n========== Page {i+1} ==========\n{text.strip()}"
            print(f"‚úÖ Page {i+1} done")

        # Save final text file
        output_path = os.path.join(output_folder, f"{base_name}.txt")
        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(full_text)

        print(f"‚úÖ Finished {filename}")

end_time = time.time()
print(f"\nüöÄ All PDFs processed in {round(end_time - start_time, 2)} seconds")
```

```python
import spacy
import os

# Load SpaCy's transformer model (more accurate) or fallback to sm model
try:
    nlp = spacy.load("en_core_web_trf")
except:
    nlp = spacy.load("en_core_web_sm")

# Folder with OCR output text
input_folder = '/content/text output'

# Entity types relevant to legal documents (you can customize)
target_labels = {"ORG", "DATE", "GPE", "LOC", "PERSON", "CARDINAL", "PRODUCT"}

# Process each file
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        print(f"\nüìÑ File: {filename}")
        with open(os.path.join(input_folder, filename), 'r', encoding='utf-8') as f:
            text = f.read()

        doc = nlp(text)

        seen = set()
        for ent in doc.ents:
            if ent.label_ in target_labels and (ent.text, ent.label_) not in seen:
                print(f"{ent.label_}: {ent.text}")
                seen.add((ent.text, ent.label_))
```
